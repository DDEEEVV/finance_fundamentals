```python
monthly_data = data.resample("M").last()
quarter_data = data.resample("Q").last()
yearly_data = data.resample("Y").last()
```

```python
data.shift(periods=5).head(10)
```

```python
data['Stock Returns'] = 1 - data['Adj Close']/ data['Adj Close'].shift(1)
```

```python
data.head()
```

```python
SMA_7 = data['Adj Close'].rolling(7).mean()
SMA_50 = data['Adj Close'].rolling(50).mean()
data['SMA_7'] = SMA_7
data['SMA_50'] = SMA_50

EMA_5 = data['Adj Close'].ewm(span=5, adjust=False).mean()
EMA_20 = data['Adj Close'].ewm(span=20, adjust=False).mean()
data['EMA_5'] = EMA_5
data['EMA_20'] = EMA_20
```
